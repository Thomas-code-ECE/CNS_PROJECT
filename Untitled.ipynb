### Find the location of the rat (x,y) with its neural activity

```python
import math
```

```python
def coordinate_to_int(df):
    df['x'] = df.apply(lambda row: int(math.ceil(row['x'])),axis = 1)
    df['y'] = df.apply(lambda row: int(math.ceil(row['y'])),axis = 1)
    return df
```

```python
def coordinate_to_zone(df,size_of_zone):
    all_zones = []
    nb_zone = max(df['x'].max(),df['y'].max())
    while(nb_zone%size_of_zone !=0):
        size_all_zone = size_all_zone+1
    df = coordinate_to_int(df)
    for coordinate in df.iterrows():
        nb_zone_x = nb_zone_y = 1
        while(size_of_zone*nb_zone_x<coordinate[1]['x']):
            nb_zone_x = nb_zone_x+1
        while(size_of_zone*nb_zone_y<coordinate[1]['y']):
            nb_zone_y = nb_zone_y+1 
        all_zones.append(str(nb_zone_x)+'x'+str(nb_zone_y))
    return all_zones
```
