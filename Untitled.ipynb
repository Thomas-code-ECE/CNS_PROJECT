### Find the location of the rat (x,y) with its neural activity

```python
import math
```

```python
def coordinate_to_int(df):
    df['x'] = df.apply(lambda row: int(math.ceil(row['x'])),axis = 1)
    df['y'] = df.apply(lambda row: int(math.ceil(row['y'])),axis = 1)
    return df
```

```python
def coordinate_to_zone(df,size_of_zone):
    all_zones = []
    nb_zone = 120
    while(nb_zone%size_of_zone !=0):
        size_of_zone = size_of_zone+1
    df = coordinate_to_int(df)
    for coordinate in df.iterrows():
        nb_zone_x = nb_zone_y = 1
        while(size_of_zone*nb_zone_x<coordinate[1]['x']):
            nb_zone_x = nb_zone_x+1
        while(size_of_zone*nb_zone_y<coordinate[1]['y']):
            nb_zone_y = nb_zone_y+1 
        all_zones.append(str(nb_zone_x)+'x'+str(nb_zone_y))
    return (all_zones,size_of_zone)
```

```python
def getDivisors(n) :
    i = 1
    divisors = []
    while i < n :
        if (n % i==0) :
            divisors.append(i)
        i = i + 1
    return divisors
```

#### Display a graph where each is colored as a function of good zone
 

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn import tree
def location_of_rat(spikes,coordinates,size_of_zone,ML_classifier):
    # Transform the float coordinates into integer 
    coordinates = coordinate_to_int(coordinates)
    # Create a new where each coordinates correspond to a zone
    coordinates['zone'],size_of_zone = coordinate_to_zone(coordinates,size_of_zone)
    # Get the number of zone created
    nb_zone = coordinates['zone'].nunique()
    print('The size of one zone is: ',size_of_zone,' and the number of zones is: ',nb_zone)
    # Split the data into train set and test set
    features_train, features_test, coordinates_train, coordinates_test = train_test_split(spikes,coordinates,train_size = 0.80)
    
    if(ML_classifier == 'LDA'):
        #LDA MODEL
        # Initialize the machine learning model
        model = LinearDiscriminantAnalysis()
        # Train the machine learning model with the train set
        model.fit(features_train,coordinates_train['zone'])
        # Determine the zone of the rat according to the spikes set by using the ML model
        zones_predicted = model.predict(features_test).tolist()
    elif(ML_classifier == 'DECISION_TREE'):
        #DECISION TREE MODEL
        # Initialize the machine learning model
        model = tree.DecisionTreeClassifier()
        # Train the machine learning model with the train set
        model.fit(features_train,coordinates_train['zone'])
        # Determine the zone of the rat according to the spikes set by using the ML model
        zones_predicted = model.predict(features_test).tolist()
    
    # Initialize the zone test set
    zones_test = coordinates_test['zone']
    # Create an array containing the coordinates (x,y) of the rat when the predicted zone is equal to the test zone
    correct_zones = [[coordinates_test.iloc[i]['x'],coordinates_test.iloc[i]['y']] for i in range(coordinates_test.shape[0]) if coordinates_test.iloc[i]['zone'] == zones_predicted[i]]
    # Create an array containing the coordinates (x,y) of the rat when the predicted zone is different to the test zone
    missed_zones = [[coordinates_test.iloc[i]['x'],coordinates_test.iloc[i]['y']] for i in range(coordinates_test.shape[0]) if coordinates_test.iloc[i]['zone'] != zones_predicted[i]]
    # Plot in green the correct predicted zone of the rat and in red the wrong predicted zone of the rat 
    plt.figure(figsize=(10, 10), dpi=150)
    if(len(correct_zones)>0):
        x_correct , y_correct = zip(*correct_zones)
    else:
        x_correct , y_correct = [],[]
    if(len(missed_zones)>0):
        x_missed , y_missed = zip(*missed_zones)
    else:
        x_missed , y_missed = [],[]
    plt.scatter(x_correct,y_correct,color = 'green',label = 'Good prediction',s = 5)
    plt.scatter(x_missed,y_missed,color = 'red',label = 'Value not found', s=5)
    if(len(correct_zones)>0):
        end_x = max(max(x_correct),max(x_missed))
        end_y = max(max(y_correct),max(y_missed))
    else:
        end_x = max(x_missed)
        end_y = max(y_missed)
    if(len(missed_zones)>0):
        end_x = max(max(x_correct),max(x_missed))
        end_y = max(max(y_correct),max(y_missed))
    else:
        end_x = max(x_correct)
        end_y = max(y_correct)
    plt.xticks(np.arange(1,end_x+size_of_zone, size_of_zone))
    plt.yticks(np.arange(1,end_y+size_of_zone, size_of_zone))
    plt.grid()
    plt.legend()
    plt.show()
    # Get the overall accuracy of the test 
    overall_accuracy = accuracy_score(zones_test,zones_predicted)
    print('The overall accuracy of the test set is equal to ',overall_accuracy)
    # Get the overall precision of the test
    overall_precision = precision_score(zones_test,zones_predicted,average='weighted',zero_division = 0)
    print('The average precision of the test set is equal to ',overall_precision)
    return [overall_accuracy,overall_precision]
```

```python
experiments_name = ['R2192','R2198','R2217','R2336','R2337']
nb_spikes = [63,33,26,48,43]
time_windows = [200,400]
divisor = []
```

```python
allMetrics = []
list_divisors = [1,2,8,20,30,60]
for index,experiment in enumerate(experiments_name):
    metrics_experiment = []
    if experiment != 'R2198':
        features_path = 'data/{}_1x{}_at{}_step200_bin100-RAW_feat.dat'.format(experiment, time_windows[0],time_windows[0]//40)
        coordinates_path = 'data/{}_1x{}_at{}_step200_bin100-RAW_pos.dat'.format(experiment, time_windows[0],time_windows[0]//40)
    else:
        features_path = 'data/{}_1x{}_at{}_step200_bin100-RAW_feat.dat'.format(experiment, time_windows[1],time_windows[1]//40)
        coordinates_path = 'data/{}_1x{}_at{}_step200_bin100-RAW_pos.dat'.format(experiment, time_windows[1],time_windows[1]//40)
    features = pd.read_csv(features_path,delim_whitespace=True,names = np.arange(1,nb_spikes[index]+1))
    coordinates = pd.read_csv(coordinates_path,delim_whitespace=True,names = ['x','y'])
    classifier = ['LDA','DECISION_TREE']
    print('Experiment '+experiment)
    print("The range of the coordinates")
    print('x:[',coordinates['x'].min(),',',coordinates['x'].max(),']')
    print('y:[',coordinates['y'].min(),',',coordinates['y'].max(),']')
    for divisor in list_divisors:
        metrics_experiment.append(location_of_rat(features,coordinates,divisor,classifier[1]))
    allMetrics.append(metrics_experiment)
```

```python
for index,experiment in enumerate(allMetrics):
    plt.figure(figsize=(8, 6), dpi=60)
    accuracy_list,precision_list = zip(*experiment)
    plt.title('Representation of the accuracy and the precision for the experiment '+experiments_name[index]+' as a function of the size of the zone')
    plt.plot(list_divisors,accuracy_list,label='accuracy',color='b')
    plt.plot(list_divisors,precision_list,label='precision',color='r')
    plt.grid()
    plt.legend()
    plt.show()
```
